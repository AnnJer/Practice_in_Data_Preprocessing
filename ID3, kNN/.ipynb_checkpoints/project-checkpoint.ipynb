```python
#stacking version

class Stacking(BaseEstimator, ClassifierMixin):
    def __init__(self, meta_model, models):
        self.models = models
        self.meta_model = meta_model
        self.n = len(models)
        self.valid = None
        
    def fit(self, X, y=None, test_size=0.3 , cv=3, err=0.001, random_state=None):
        if(test_size>0):
            train, valid, y_train, y_valid = train_test_split(X, y, test_size = test_size, random_state=random_state)
            
            self.val = np.zeros((val.shape[0], self.n))
            for i, clf in enumerate(self.models):
                clf.fit(train, y_train)
                self.valid[:, i] = clf.predict(valid)
            
            self.meta_model.fit(self.valid, y_valid)
        
        else:
            self.valid = err*np.random.randn(X.shape[0], self.n)
            
            for i, clf in enumerate(self.models):
                self.valid[:, i] += cross_val_predict(clf, X, y, cv=cv, n_jobs=-1, method='predict')
                clf.fit(X, y)
            
            self.ens_model.fit(self.valid, y) 
            
        return self
    
    def predict(self, X, y=None):
        X_meta = np.zeros((X.shape[0], self.n))
        
        for t, clf in enumerate(self.models):
            X_meta[:, t] = clf.predict(X)
        
        c = self.ens_model.predict(X_meta)
        
        return (c)
```

```python

data= pd.read_csv('/home/ann/na_git/ID3, kNN/data.csv', header=None, 
                  names=['Variance', 'Skewness', 'Curtosis', 'Entropy', 'Class'])

y=data['Class']
X = data.iloc[:, :4].values
# X.head()
# preprocessing.MinMaxScaler(X)

train_X,  test_X, train_y, test_y = train_test_split(X, y, train_size=0.7, random_state=42)


# plt.figure(figsize=(12, 10))
# plt.scatter(train_X['Variance'], train_X['Entropy'], c=train_y, s=20, alpha=0.5, label='train')
# plt.scatter(test_X['Variance'], test_X['Entropy'], c=test_y, s=5, alpha=1.0, label='test')
# plt.xlabel('Variance')
# plt.ylabel('Entropy')
# plt.legend()
# plt.show()

```

```python


knn1 = KNeighborsRegressor(n_neighbors=3)
knn1.fit(train_X, train_y)
launching_plotting(knn1, test_X, test_y, '3NN')

knn2 = KNeighborsRegressor(n_neighbors=10)
knn2.fit(train_X, train_y)
launching_plotting(knn2, test_X, test_y, '10NN')


rg0 = Ridge(alpha=0.01)
rg0.fit(train_X, train_y)
launching_plotting(rg0, test_X, test_y, 'ridge-0.01')

rg1 = Ridge(alpha=1.1)
rg1.fit(train_X, train_y)
launching_plotting(rg1, test_X, test_y, 'ridge-1.1')

rg2 = Ridge(alpha=100.1)
rg2.fit(train_X, train_y)
launching_plotting(rg2, test_X, test_y, 'ridge-100.1')


rf1 = RandomForestRegressor(n_estimators=100, max_depth=1)
rf1.fit(train_X, train_y)
launching_plotting(rf1, test_X, test_y, 'rf-d1')

rf2 = RandomForestRegressor(n_estimators=100, max_depth=5)
rf2.fit(train_X, train_y)
launching_plotting(rf2, test_X, test_y, 'rf-d5')


gbm1 = lgb.LGBMRegressor(boosting_type='gbdt', learning_rate=0.05, max_depth=2, n_estimators=200, nthread=-1, objective='regression')    
gbm1.fit(train_X, train_y)
launching_plotting(gbm1, test_X, test_y, 'gbm-d2')

gbm2 = lgb.LGBMRegressor(boosting_type='gbdt', learning_rate=0.05, max_depth=5, n_estimators=200, nthread=-1, objective='regression')    
gbm2.fit(train_X, train_y)
launching_plotting(gbm2, test_X, test_y, 'gbm-d5')
```

```python
import kmapper as km
from IPython.core.display import display, HTML
from selenium import webdriver
driver = webdriver.Firefox()


mapper = km.KeplerMapper(verbose=2)

projected_data = mapper.fit_transform(X, projection=TSNE())

graph = mapper.map(projected_data,
                   clusterer=DBSCAN(eps=0.3, min_samples=3),
                   nr_cubes=10, overlap_perc=0.25)

display(mapper.visualize(graph,
                 path_html=""))
driver.get('mydata2.html')
```
